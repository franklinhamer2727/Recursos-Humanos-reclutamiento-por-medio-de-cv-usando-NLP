```python
import os
from tqdm import tqdm
from pdf2image import convert_from_path

def convertir_pdf_to_image(pdf_path, image_prefix):
    try:
        images = convert_from_path(pdf_path, dpi=300)
        # Configurar la barra de progreso
        progress_bar = tqdm(total=len(images), desc="Convirtiendo páginas")
        for i, image in enumerate(images):
            image_path = f"{image_prefix}_page{i + 1}.jpg"
            image.save(image_path, 'JPEG')
            # Actualizar la barra de progreso
            progress_bar.update(1)
        # Cerrar la barra de progreso
        progress_bar.close()
    except Exception as e:
        print(f"Error al convertir el archivo: {pdf_path}")
        print(f"Mensaje de error: {str(e)}")


```

```python
def ejecutar_convertir_pdf_to_image(ruta_base_input, ruta_base_output):
    for archivo in tqdm(os.listdir(ruta_base_input)):
        if archivo.endswith('.pdf'):
            nombre_archivo = os.path.splitext(archivo)[0]
            carpeta_archivo = os.path.join(ruta_base_output, nombre_archivo) +'/'
            os.makedirs(carpeta_archivo, exist_ok=True)
            ruta_input = os.path.join(ruta_base_input, archivo)
            convertir_pdf_to_image(ruta_input, carpeta_archivo)
```

```python
ruta_base_input = "/home/code/Escritorio/Tesis/Recursos-Humanos-reclutamiento-por-medio-de-cv-usando-NLP/data/curriculum_vitae_data/pdf/"
ruta_base_output = "/home/code/Escritorio/Tesis/Recursos-Humanos-reclutamiento-por-medio-de-cv-usando-NLP/data/curriculum_vitae_data/image/"
inicio = time.time()
ejecutar_convertir_pdf_to_image(ruta_base_input, ruta_base_output)
fin = time.time()
```
